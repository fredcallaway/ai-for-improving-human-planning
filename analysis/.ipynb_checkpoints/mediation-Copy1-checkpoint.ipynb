## Bootstrapping

### Means

```python
N_BOOT = 10000
test_score = pdf.set_index('feedback').test

def ci(xs):
    return np.quantile(xs, [0.025, 0.975])

def boot_means(n=N_BOOT):
    r = {}
    for fb in fb_order:
        x = test_score[fb]
        means = [x.sample(frac=1, replace=True).mean() for _ in range(n)]
        r[fb] = np.array(means)
    return r
```

```python
bm = boot_means()
for fb in fb_order:
    x = test_score[fb].mean()
    a, b = ci(bm[fb])
    if fb == 'none':
        tex = (rf'${x:.2f}$ points/trial (95\% CI: [${a:.2f}$, ${b:.2f}$])')
    else:
        tex = (rf'${x:.2f}$ points/trial; 95\% CI: [${a:.2f}$, ${b:.2f}$]')
    write_tex(f'mean-{fb}', tex)
```

### Differences by permutation test

```python
def null_mean_diff_samples(c1, c2, n=N_BOOT):
    r = []
    x = test_score.copy()
    for i in range(n):
        np.random.shuffle(x.values)
        r.append(x[c1].mean() - x[c2].mean())
    return np.array(r)

#         x = test_score.sample(n=n_fb[c1], replace=True)
#         y = test_score.sample(n=n_fb[c2], replace=True)
#         r.append(x.mean() - y.mean())

def report(c1, c2, n=N_BOOT):
    actual_diff = test_score[c1].mean() - test_score[c2].mean()
    null_diff = null_mean_diff_samples(c1, c2, n=n)
#     p = (1 + np.sum(abs(null_diff) >= abs(actual_diff))) / (n+1)
    p = np.mean(abs(null_diff) >= abs(actual_diff))
    write_tex(f'permutation-{c1}-{c2}', f'${pval(p)}$')

report('meta', 'none')
if 'action' in fb_order:
    report('meta', 'action')
    report('action', 'none')
```

## Mediation

```python
rdf = test_block.copy()
# rdf['feedback'] = (rdf.feedback != 'none').astype(int)
fb = rdf.pop('feedback')
for c in ['action', 'meta']:
    rdf[c] = (fb == c).astype(int)

rdf['clicked'] = rdf.clicked.astype(int)
rdf = rdf[['action', 'meta', 'backward', 'clicked', 'n_clicks', 'score']].reset_index()

GROUP_LEVEL = True
if GROUP_LEVEL:
    rdf.backward.fillna(0, inplace=True)
    rdf = rdf.groupby('pid').mean()
else:
    rdf.backward.fillna(0, inplace=True)
    for k in 'action meta backward clicked'.split():
        rdf[k] = rdf[k].astype(int)
```

```python
%%R -i rdf
library(lavaan)
model <- ' 
# direct effects
score ~ sm * meta
score ~ action

score ~ sb * backward
score ~ sc * clicked
score ~ sn * n_clicks

clicked ~ cm * meta
clicked ~ ca * action

backward ~ bm * meta
backward ~ ba * action

n_clicks ~ nm * meta
n_clicks ~ na * action

score_backward_meta := sb * bm
score_clicked_meta := sc * cm
score_nclicks_meta := sn * nm
score_backward_action := sb * ba
score_clicked_action := sc * ca
score_nclicks_action := sn * na

total_meta := score_backward_meta + score_clicked_meta + score_nclicks_meta + sm
'
fit <- sem(model, data=rdf)
summary(fit, fit.measures=TRUE)
```

```python
est = estimate.query('op == "~"').copy()
est['parameter'] = est.pop('lhs') + ' ~ ' + est.pop('rhs')
for i, row in est.iterrows():
    name = row.parameter.replace(' ~ ', '-').replace('n_clicks', 'nclicks')
    lo, hi = row['ci.lower'], row['ci.upper']
    write_tex(name, rf'${row.est:.2f}$')
    
    write_tex(name + '-ci', rf'95\% CI: [{lo:.2f}, {hi:.2f}], ${pval(row.pvalue)}$')
```
