```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_context('notebook', font_scale=1.3)
sns.set_style('white')
```

```python
from analysis_utils import *
data = get_data('M2.0', '../data/human/')

df = data['maze']
df.rename(columns={'name': 'maze', 'reverse': 'forward_easy'}, inplace=True)
df['time'] = df.events.apply(lambda x: x[-1]['time']) / 1000
df['start_time'] = df.events.apply(lambda x: x[0]['time']) / 1000
# df['trial'] = np.tile(np.arange(8), len(df.pid.unique())) + 1

from toolz.curried import get
pdf = data['participants']
pdf['time'] = df.groupby('pid').time.mean()
survey = data['survey-text']

events = df.events[0]
def loc(event):
    kind, x, y = event['event'].split(' ')
    return (int(x), int(y))
df['clicks'] = df.events.apply(lambda events: list(map(loc, events)))
df['first_click'] = df.clicks.apply(get(0))

def dist(x, y):
    return abs(x[0] - y[0]) + abs(x[1] - y[1])

def distances(row):
    return [dist(row.start, click) - dist(row.goal, click) for click in row.clicks]

df['distances'] = df.apply(distances, axis=1)
df['planned_backward'] = df.distances.apply(lambda x: sum(x[:10]) > 0)
df.planned_backward.mean()
```
